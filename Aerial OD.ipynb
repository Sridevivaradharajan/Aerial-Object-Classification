```python

import os
import yaml
from pathlib import Path
import shutil
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
# ==========================================
# 1. DATASET PREPARATION & CONFIGURATION
# ==========================================

def create_data_yaml(dataset_path, output_path='data.yaml'):
    """
    Create data.yaml configuration file for YOLOv8
    """
    data_config = {
        'path': str(Path(dataset_path).absolute()),  # Dataset root directory
        'train': 'train/images',  # Training images path
        'val': 'valid/images',    # Validation images path
        'test': 'test/images',    # Test images path
        
        'nc': 2,  # Number of classes
        'names': ['bird', 'drone']  # Class names (class_id 0 and 1)
    }
    
    with open(output_path, 'w') as f:
        yaml.dump(data_config, f, sort_keys=False)
    
    print(f"âœ… data.yaml created at: {output_path}")
    return output_path
```
